# Program to download videos from youtube

```python
from pytube import YouTube

#where to save
SAVE_PATH = "D:/" #to_do
link=["https://www.youtube.com/watch?v=OBHtu285aqE"]
for i in link:
    yt = YouTube(i)
    #Title of video
    print("Title: ",yt.title)
    #Number of views of video
    print("Number of views: ",yt.views)
    #Length of the video
    print("Length of video: ",yt.length,"seconds")
    # print("Description: ",yt.description)
    #print("Ratings: ",yt.rating)
    ys = yt.streams.get_highest_resolution()
    #ys = yt.streams.get_by_itag('137')
    ys.download(SAVE_PATH)
```

# Program to convert a video to multiple images

```python
import cv2
import numpy as np
import os

# set video file path of input video with name and extension
vid = cv2.VideoCapture('D://Log Analysis with Splunk  How to use Splunk to analyse a Real time Log  Splunk Use Cases  Edureka.mp4')


if not os.path.exists('images'):
    os.makedirs('images')

#for frame identity
index = 0
while(True):
    # Extract images
    ret, frame = vid.read()
    # end of frames
    if not ret: 
        break
    # Saves images
    name = './images/frame' + str(index) + '.jpg'
    print ('Creating...' + name)
    cv2.imwrite(name, frame)

    # next frame
    index += 1
```
